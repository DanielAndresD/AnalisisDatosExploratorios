# Creacion de un DataSet

## Importar Librerías

```python
import pandas as pd
```

## Definir rutas de los archivos

```python
df_path="../datasets/DatosSeguros.csv"
```

## Crear el dataset

```python
df=pd.read_csv(df_path)
df
```

## Analizar la estructura del dataset

```python
df.info()
```

Contar valores nulos

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.info()
```

```python
df=df.drop_duplicates()
df.info()
```

### Ordenar data 

```python
df.sort_values('edad')
```

```python
df.sort_values('edad',ascending=False)
```

## Agrupar datos

```python
rangoEdad=[17,28,38,48,58,68]
nombreRangoEdad=['A (17-28)','B (29-38)','C (39-48)','D (49-58)','E (59-68)']
```

```python
# Crear nueva columna
df['rango_edad']=pd.cut(df['edad'],rangoEdad,labels=nombreRangoEdad)
df
```
